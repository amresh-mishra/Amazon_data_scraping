```python
import time
import pandas as pd
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

options = webdriver.ChromeOptions()
options.add_argument('--no-sandbox')
options.add_argument("--log-level=3")
user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36"
options.add_argument(f"user-agent={user_agent}")

driver = webdriver.Chrome(options=options)
wait = WebDriverWait(driver, 10)

driver.get("https://www.amazon.in")
keyword = input("Enter the keyword: ")

search_box = driver.find_element(By.ID, 'twotabsearchtextbox')
search_box.send_keys(keyword)

search_button = driver.find_element(By.ID, 'nav-search-submit-button')
search_button.click()

# Define empty lists to store data
product_name = []
product_price = []
product_ratings = []
product_ratings_num = []
product_description = []

# Take input from the user about how many pages to scroll
num_pages_to_scroll = int(input("Enter the number of pages you want to scroll: "))

for _ in range(num_pages_to_scroll):
    # Use WebDriverWait consistently for waiting
    items = wait.until(EC.presence_of_all_elements_located((By.XPATH, '//div[contains(@class, "s-result-item s-asin")]')))
    
    for item in items:
        title_link = item.find_element(By.XPATH, './/span[contains(@class, "a-size-medium a-color-base a-text-normal") or contains(@class, "a-size-base-plus a-color-base a-text-normal")]')
        
        # Scroll to the title link
        driver.execute_script("arguments[0].scrollIntoView();", title_link)
        
        # Wait for overlay/loading to disappear
        wait.until(EC.invisibility_of_element_located((By.ID, 'loading-indicator')))  # Replace 'loading-indicator' with appropriate selector
        
        title_link.click()

        # Switch to the new window
        driver.switch_to.window(driver.window_handles[-1])

        # Extract the description from the new window
        description_element = wait.until(EC.presence_of_element_located((By.CLASS_NAME, 'a-unordered-list.a-vertical.a-spacing-mini')))
        description = description_element.text if description_element else "Description not available"
        product_description.append(description)

        # Close the new window
        driver.close()

        # Switch back to the original window
        driver.switch_to.window(driver.window_handles[0])

        name = item.find_element(By.XPATH, './/span[contains(@class, "a-size-medium a-color-base a-text-normal") or contains(@class, "a-size-base-plus a-color-base a-text-normal")]')
        if name:
            product_name.append(name.text)
        else:
            product_name.append("Name not available")

        whole_price = item.find_elements(By.XPATH, './/span[@class="a-price-whole"]')

        if whole_price:
            product_price.append(whole_price[0].text)
        else:
            product_price.append("Price not available")

        ratings_box = item.find_elements(By.XPATH, './/div[@class="a-row a-size-small"]/span')

        if ratings_box:
            ratings = ratings_box[0].get_attribute('aria-label')
            ratings_num = ratings_box[1].get_attribute('aria-label')
        else:
            ratings, ratings_num = "0", "0"

        product_ratings.append(ratings)
        product_ratings_num.append(ratings_num)

    # Locate and interact with the next page button
    next_page_button = driver.find_element(By.CSS_SELECTOR, '.s-pagination-next')
    if 's-disabled-pagination-next' in next_page_button.get_attribute('class'):
        print("No more pages to scrape.")
        break
    
    # Scroll to the next page button and click it
    driver.execute_script("arguments[0].scrollIntoView();", next_page_button)
    next_page_button.click()

    # Wait for the new search results to load
    wait.until(EC.staleness_of(items[0]))

driver.quit()

data = {
    "Product Name": product_name,
    "Product Price": product_price,
    "Product Ratings": product_ratings,
    "Product Ratings Numbers": product_ratings_num,
    "Product Descriptions": product_description
}

df = pd.DataFrame(data)

# Save the DataFrame to a CSV file
csv_filename = f"{keyword}_search_results.csv"
df.to_csv(csv_filename, index=False)

print(f"Data saved to {csv_filename}")
```
